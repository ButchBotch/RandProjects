```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#Resize graph
plt.figure(figsize=(5,3),dpi=1000)
#plt.plot(x,y,label='2x',color='red',linewidth=5,linestyle='--',marker='.',markersize='20',markeredgecolor='blue') # plt.plot(x,y)
plt.plot(x,y,'b^--',label='2x')

x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:4],x2[:4]**2,'r',label='x^2')
plt.plot(x2[3:],x2[3:]**2,'g--')

plt.title('first graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('manes')
plt.ylabel('ylabel')

zts=np.arange(0,17,2)
print(zts)
plt.xticks([0,2,4])
plt.yticks(zts)

plt.legend()

plt.savefig('savedgraph.png')
plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head(len(fifa))
```

```python
fig=plt.figure(dpi=300)
bins=np.arange(40,101,10)
plt.hist(fifa.Overall,bins=bins,color='magenta') # or google color picker
plt.xticks(bins)
plt.ylabel('number of players')
plt.xlabel('Skill level')
plt.title('Distribution of Player Skills in FIFA 2018')
#plt.yscale("log")
plt.show()
```

```python
# Pie chart, left right footed
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
other=fifa.loc[(fifa['Preferred Foot']!='Right')&(fifa['Preferred Foot']!='Left')].count()[0]
#print('left',left,'right',right,'neither',other)
#print('all :',len(fifa),'left+right+other :',left+right+other,'right+non-right :', right+other)
plt.figure(dpi=300)
labels=['Left','Right','Other']
colors=['Blue','Yellow','Red']
plt.pie([left,right,other],labels=labels,colors=colors,autopct='%.2f %%')
plt.title('Foot Preference of Fifa Players')
plt.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]

plt.style.use('ggplot')
weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']
explode=(.4,.2,.0,.0,.2)

plt.figure(dpi=300)
plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

```python
barcelona=fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
realm=fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs=fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels=['Barca','Real Madrid','New England Revolut']

plt.style.use('default')
plt.figure(figsize=(6,6),dpi=300)
boxes=plt.boxplot([barcelona,realm,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})

count=0
tricolor=['red','green','blue']
for box in boxes['boxes']:
    box.set(color=tricolor[count])
    box.set(facecolor='#e0e0e0')
    count+=1
    
plt.ylabel('Overall Rating')
plt.title('Professional Football Team Comparison')

plt.show()
```
