```python
sns.catplot(x='survived',y='age',hue='sex',kind='swarm',data=titanic)
sns.catplot(x='survived',y='age',hue='pclass',kind='swarm',data=titanic)
```

```python
# Drop all rows with NAN
nan_df=titanic[titanic.isna().any(axis=1)]
titanic2=titanic.dropna(subset=['survived','age','pclass']) 
# any:for any single nan , all: for whole rows of NaN
sns.catplot(x='survived',y='age',hue='sex',kind='swarm',data=titanic2)
#sns.distplot(titanic['age']);
len(titanic2['survived'])
```

```python
titanic2['woman_bin'] = np.where(titanic2['sex']=='female' , 1, 0)
titanic2['15_ol_bin'] = np.where(titanic2['age']<=15 , 1, 0)
titanic2['15_ol_or_woman']=titanic2['woman_bin']+titanic2['15_ol_bin']
titanic2['woman_or_under15']=np.where(titanic2['15_ol_or_woman']>=1 , 1, 0)
titanic2['pc1'] = np.where(titanic2['pclass']==1 , 1, 0)
titanic2['woman_or_under15_or_pc1']=np.where(titanic2['woman_or_under15']+titanic2['pc1']>=1 , 1, 0)
titanic2.head(40)
```

```python
 sns.catplot(x='survived',y='age',hue='woman_or_under15',kind='swarm',data=titanic2)
 sns.catplot(x='survived',y='age',hue='woman_or_under15_or_pc1',kind='swarm',data=titanic2)
 sns.catplot(x='survived',y='age',hue='pc1',kind='swarm',data=titanic2)
```

```python
survival_rate_general=titanic2['survived'].sum()/len(titanic2['survived'])

pc1=titanic2.groupby('pc1').sum()#['survived']
pc1_rate=pc1['survived'][1]/titanic2['survived'].sum()
pc1_rate_g=pc1['survived'][1]/len(titanic2['survived'])

woman=titanic2.groupby('sex').sum()#['survived']
woman_rate=woman['survived'][0]/titanic2['survived'].sum()
woman_rate_g=woman['survived'][0]/len(titanic2['survived'])

woman_or_under15=titanic2.groupby('woman_or_under15').sum()#['survived']
woman_or_under15_rate=woman_or_under15['survived'][1]/titanic2['survived'].sum()
woman_or_under15_rate_g=woman_or_under15['survived'][1]/len(titanic2['survived'])

woman_or_under15_or_pc1=titanic2.groupby('woman_or_under15_or_pc1').sum()#['survived']
woman_or_under15_or_pc1_rate=woman_or_under15_or_pc1['survived'][1]/titanic2['survived'].sum()
woman_or_under15_or_pc1_rate_g=woman_or_under15_or_pc1['survived'][1]/len(titanic2['survived'])


print('               Survival_rate :',survival_rate_general)
print('\n                     class 1 :',pc1_rate_g)
#print('                   class 2-3 :',survival_rate_general-pc1_rate_g)
print('                      female :',woman_rate_g)
#print('                        male :',survival_rate_general-woman_rate_g)
print('            woman or under15 :',woman_or_under15_rate_g)
#print('                 man over 15 :',survival_rate_general-woman_or_under15_rate_g)
print('woman or under 15 or pclass1 :',woman_or_under15_or_pc1_rate_g)

print('\n           Of those survived ')
print('                      class1 :',pc1_rate)
print('                       woman :',woman_rate)
print('            woman or under15 :',woman_or_under15_rate)
print('woman or under 15 or pclass1 :',woman_or_under15_or_pc1_rate)
```
