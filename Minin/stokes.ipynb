```python
import yahoofinance as yf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
historical = yf.HistoricalPrices('AAPL', '2010-01-01', '2020-03-10')
dfs = historical.to_dfs()
df = dfs['Historical Prices']
```

```python
df.index = pd.to_datetime(df.index)
```

```python
fig1 = plt.figure(figsize=(15,10),dpi=400)
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('AAPL Share Price')
plt.plot(df[['Close']])
plt.grid(c='grey')
plt.xticks(rotation=45)
```
